```julia
using Plots, Measures
```

```julia
xGridD = 0:2
pDiscrete = [0.25, 0.25, 0.5];
```

```julia
xGridC = -1:0.01:1
pContinuous(x) = 3/4*(1 - x^2)
pContinuous₂(x) = x < 0 ? x+1 : 1-x ;
```

```julia
p₁ = plot(xGridD, line=:stem, pDiscrete, marker=:circle, color=:blue, ms=6, msw=0)
p₂ = plot(xGridC, pContinuous.(xGridC), color=:blue)
p₃ = plot(xGridC, pContinuous₂.(xGridC), color=:blue)
plot(p₁, p₂, p₃
    , layout=(1,3), legend=false
    , ylims=(0,1.1)
    , xlabel="x", ylabel=["Probability" "Density" "Density"]
    , size=(1200, 400)
    , margin=5mm
)
```
